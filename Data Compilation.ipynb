We now have a list of lists, the first list is a list of the headers and each following list corresponds to a row from the csv


Now we clean up the data.
We want to reformat this so that there are 5 columns: 
    - Year 
    - Neighbourhood ID (N_ID) 
    - Neighbourhood Name (Location) 
    - Population (Pop) 
    - Total Crime Rate (C_Rate) 
        Where total crime rate is the sum of all the different crime rates in a specific year, divided by the number of different types of crime, per neighbourhood. 
We will call this new data frame crate_df.

We need to identify the indices of the columns we are keeping, and input corresponding row elements into a new list of lists.

```python
indices = []
rows = []
list_of_csv[0]
names = ['HoodID', 'Rate', 'Pop']
```

```python
for i in range(len(list_of_csv[0])):
    for name in names:
        if name in list_of_csv[0][i]:
                   indices.append(i)         
               
```

```python
headers = ['Year', 'N_ID',  'Pop', 'C_Rate']
```

```python
ids = []
for i in range(len(list_of_csv)-1):
    if list_of_csv[i+1][3] not in ids:
        ids.append(int(list_of_csv[i+1][3]))
max(ids)
#140 different naighbourhoods
```

```python
#defining a function to calculate the crime rate given a neighbourhood ID and a year
def crate(n,y):
    n = str(n)
    y = str(y)
    rates = []
    for i in range(len(list_of_csv)):
        #iterate through each row
        for j in range(len(list_of_csv[i])):
            #iterate through each element of each row
            if list_of_csv[i][3] == n:
                #if the selected rows neighbourhood id matches our input id
                if (y in list_of_csv[0][j]) and (j in indices):
                    #if the element corresponds to a stat from our input year
                    if list_of_csv[i][j] == '':
                        rates.append(0)
                    else:                    
                        rates.append(float(list_of_csv[i][j]))
                    #add the stat to our rates list.
    if len(rates) != 0:
        avg = sum(rates)/len(rates)
    else:
        avg = 0
    
    return avg
                                
```

```python
def find_pop(n,y):
    for i in range(len(list_of_csv)):
        if list_of_csv[i][3] == str(n):
            a = list_of_csv[i][y - 2009:len(list_of_csv):16]
    for j in range(len(a)):
        if a[j] == '':
            a[j] = 0
        else:
            a[j] = float(a[j])
    total_crime = sum(a)
    for k in range(len(new_data)):
        for l in range(len(new_data[k])):
            if (new_data[k][0] == y) and (new_data[k][1]==n):
                rate = new_data[k][3]
    pop = total_crime/(rate*len(a))*100000
    return pop
```
